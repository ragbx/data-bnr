```python
med_pra = az[az['unit'] == 'MED_PRA']
med_pra.groupby(['subunit', 'folder']).aggregate({'name':'size', 'size_mo': 'sum'}).reset_index()
```

```python
med_pra.groupby(['folder'])['name'].count()
```

```python
med_pra.pivot_table(index=['folder', 'unit'], values=['name', 'size'], aggfunc={'name':'count', 'size':'sum'})
```

```python
med_pra.pivot_table(index=['unit', 'sfolder'], values=['name', 'size'], aggfunc={'name':'size', 'size':'sum'})
```
